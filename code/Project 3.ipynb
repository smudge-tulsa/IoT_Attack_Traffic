```python
# rename columns for better clarity
tfc.rename(columns={'id.orig_p':'origin port','id.resp_p':'response port','Attack_type':'traffic_pattern'},inplace=True)
# save column names to a list
tfcCols = tfc.columns.tolist()
tfcCols
```

```python
# 'traffic_patterns' are the benign and attack traffic present in the data
# identify which ones are present and save to a df
patterns=tfc.traffic_pattern.unique()
tfcPatterns=pd.DataFrame(patterns, columns=['pattern'])
```

```python
tfcPatterns
```

```python
# use a function to search throught the pattern values to determine if each is an attack or benign
def categorize_attack(pattern):
    if pattern.startswith('NMAP'):
        return 'attack'
    elif 'DOS' in pattern:
        return 'attack'
    elif 'Metasploit' in pattern:
        return 'attack'
    elif pattern.startswith('ARP'):
        return 'attack'
    else:
        return 'benign'
# apply the function to populate thew new attack? column to the patterns df
tfcPatterns['attack?'] = tfcPatterns['pattern'].apply(categorize_attack)
tfcPatterns
```
